# Make density plots

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Inputs

```python
results_df = pd.read_csv("input_files/cryoET_density_df.csv",
                        index_col = 0)

# Radius, within which neighbors are counted
# Based on diameter of cohesin - dockerin complex + average domain size
dist_cutoff_nm = 11
```

## Plot

```python
# Carb Palette defined based on ColorBrewer 5-class BrBG
carb_palette = {"Fructose": "#969696",
                "RS": "#018571"}
```

### For paper

```python
inch = 2.54 # one inch is 2.54 cm

fig, ax = plt.subplots(figsize=(6 / inch,
                                6 / inch))

sns.histplot(results_df,
            x = "dist_cw_nm",
            hue = "condition",
            palette = carb_palette,
            stat = "percent",
            ax = ax,
            legend = True,
            common_norm = False,
            bins = 20,
             #binrange = [0,200],
             kde=True,
             multiple='layer',
            )

#ax.set_xlim(0,200)
ax.set_xticklabels(labels = ax.get_xticklabels(),
                   fontsize = 8)
ax.set_xlabel('Distance to Cell Wall [nm]',
              fontsize = 8)

ax.set_yticklabels(labels = ax.get_yticklabels(),
                   fontsize = 8)
ax.set_ylabel("Percent",
              fontsize = 8)

ax.legend(['RS', 'Fructose'],
    title = (),
          fontsize = 5,
          frameon = False)
#sns.move_legend(ax, 'upper left', bbox_to_anchor=(1,0.5),
#                title = None, frameon = False,)

fig.tight_layout(pad=1)
```
